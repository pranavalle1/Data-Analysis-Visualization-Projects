```python
# 1)Show any one data formatting technique on numeric as well as categorical data
```

```python
obj_columns = df.select_dtypes(include=np.float).columns.tolist()
df[obj_columns] = df[obj_columns].astype('int')
print(df)
```

```python
#data formatting on categorical data 
df['Black'] = df['Black'].str.upper()
print(df)
```

```python
#data formatting on numerical data 
df['GameNumber'] = df['GameNumber'].apply(np.floor)
print(df)
```

```python
# 2) Show any one Data Normalisation technique using on numeric as well as categorical data
# Normalising Numeric data using single feature scaling
```

```python
#data normalization technique on numeric
from sklearn.preprocessing import MinMaxScaler
X = np.array(df['GameNumber']).reshape(-1,1)
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
df['min max'] = X_scaled.reshape(1,-1)[0]

df.head()
```

```python
#data normalization technique on categorical
one_hot_encoded_df = pd.get_dummies(df, columns = ['Black'])
print(one_hot_encoded_df)
```

```python
# 3) Demonstrate the use of lambda anonymous function
df['GameNumber'] = df['GameNumber'].apply(lambda x: np.log(x) if x != 0 else 0)
print(df)
```

```python
# 4.Correlation
from scipy.stats import pearsonr
def get_correlation(column1, column2, df):
    pearson_corr, p_value = pearsonr(df[column1], df[column2]) 
    print("Correlation between {} and {} is {}".format(column1, column2, pearson_corr))
    print("P-value of this correlation is {}".format(p_value))
```
